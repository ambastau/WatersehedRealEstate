````python
# Average Nightly Rent for short term rental in different cities.
````

````python
%%sql
SELECT AVG(s.sample_nightly_rent_price) as Average_Rent, l.city
FROM st_rental_prices as s, location as l, watershed_property_info as w
WHERE w.location = s.location and w.property_type = s.property_type and w.location = l.location_id
GROUP BY l.city
ORDER BY Average_Rent;
````

````python
# Average Nigtly Price of 1 and 2 bedroom property
````

````python
%%sql
SELECT AVG(s.sample_nightly_rent_price) as AveragePrice, p.num_bedrooms
FROM st_rental_prices as s, watershed_property_info as w, property_type as p
WHERE s.location = w.location and w.property_type = s.property_type and s.property_type = p.property_type_id
GROUP BY p.num_bedrooms;
````

````python
# Extracting Average Rental Price, city and property type
````

````python
%%sql
SELECT AVG(s.sample_nightly_rent_price) as AverageRent, l.city, p.apt_house
FROM st_rental_prices s
INNER JOIN watershed_property_info w
ON s.location = w.location and s.property_type = w.property_type
INNER JOIN location l
ON w.location = l.location_id
INNER JOIN property_type p
ON s.property_type = p.property_type_id
GROUP BY l.city, p.apt_house;
````
